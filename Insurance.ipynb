```python
from tkinter import Tk, Label, Entry, Button
import joblib

def show_entry():
    try:
        p1 = float(age_entry.get())
        p2 = float(gender_entry.get())
        p3 = float(bmi_entry.get())
        p4 = float(children_entry.get())
        p5 = float(smoker_entry.get())
        p6 = float(region_entry.get())

        model = joblib.load('model_joblib_gr')
        result = model.predict([[p1, p2, p3, p4, p5, p6]])

        Label(master, text="Insurance Cost").grid(row=7, column=0)
        Label(master, text=f"${result[0]:.2f}").grid(row=7, column=1)

    except Exception as e:
        Label(master, text="Error: Invalid Input", fg="red").grid(row=7, columnspan=2)

# Create the main window
master = Tk()
master.title("Insurance Cost Prediction")

# Title Label
Label(master, text="Insurance Cost Prediction", bg="black", fg="white").grid(row=0, columnspan=2)

# Input Labels
Label(master, text="Enter Your Age").grid(row=1, column=0)
Label(master, text="Male or Female [1/0]").grid(row=2, column=0)
Label(master, text="Enter Your BMI").grid(row=3, column=0)
Label(master, text="Number of Children").grid(row=4, column=0)
Label(master, text="Smoker Yes/No [1/0]").grid(row=5, column=0)
Label(master, text="Region [1-4]").grid(row=6, column=0)

# Entry Fields
age_entry = Entry(master)
age_entry.grid(row=1, column=1)

gender_entry = Entry(master)
gender_entry.grid(row=2, column=1)

bmi_entry = Entry(master)
bmi_entry.grid(row=3, column=1)

children_entry = Entry(master)
children_entry.grid(row=4, column=1)

smoker_entry = Entry(master)
smoker_entry.grid(row=5, column=1)

region_entry = Entry(master)
region_entry.grid(row=6, column=1)

# Button to Predict
predict_button = Button(master, text="Predict", command=show_entry)
predict_button.grid(row=7, column=0, columnspan=2, pady=10)

# Run the application
master.mainloop()
```
